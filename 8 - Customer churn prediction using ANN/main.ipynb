```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
data
```

```python
data.drop('customerID',axis=1,inplace=True)
data
```

```python
tenure_churn_yes = data[data.Churn=='Yes'].tenure
tenure_churn_no = data[data.Churn=='No'].tenure

plt.hist([tenure_churn_yes,tenure_churn_no],color=['green','red'],label=['Churn yes','Churn no'])
plt.title('Çıktı ve Tenuture Sütunlarına Göre Bar Grafiği')

plt.xlabel('Tenure')
plt.ylabel('Number of customers')

plt.legend()
plt.show()
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(data)
```

```python
data.replace('No internet service','No',inplace=True)
data.replace('No phone service','No',inplace=True)
print_unique_col_values(data)
```

```python
data[pd.to_numeric(data['TotalCharges'], errors='coerce').isnull()]
```

```python
data = data[data['TotalCharges'] != ' ']
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
data.info()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['Churn'],axis=1)
y= data['Churn']
```

```python
X = pd.get_dummies(data=X)# ,drop_first=True
y= pd.get_dummies(y,drop_first=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
"""
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

or

import tensorflow as tf

X_train = tf.convert_to_tensor(X_train)
y_train = tf.convert_to_tensor(y_train)

"""
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
from tensorflow.keras.layers import Dense,InputLayer
from tensorflow.keras import Sequential
```

```python
model = Sequential([
    Dense(100,input_dim=X_train.shape[1],activation='relu'),
    Dense(3000,activation='relu'),
    Dense(1000,activation='relu'),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
yp[:,0]
```

```python
yp =[round(i) for i in yp[:,0]]
yp
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,yp))
```

```python
import seaborn as sn
from tensorflow.math import confusion_matrix
cm = confusion_matrix(labels=y_test,predictions=yp)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
